```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# conversion 
HA_TO_KJ_MOL = 2625.49962

# crystal energy
PBE_D3BJ = "benzene_crystal_pbe_d3bj_energy.csv"
pbe_d3bj_df = pd.read_csv(PBE_D3BJ)

functional_list = {}
functional_list["pbe_d3bj"] = pbe_d3bj_df

# molecular energy with counterpoise correction. The ordering is DZ, TZ, QZ
molecular_energies = {}
molecular_energies["pbe_d3bj"] = np.array([-231.9655513248, -232.0268653837, -232.0435139637])
```

```python
func_list = ["pbe_d3bj"]

for func in func_list:
    print(func)
    data = functional_list[func]
    crystal_dz_energy = data["DZ"].astype(float).to_numpy() #DZ 
    crystal_tz_energy = data["TZ"].astype(float).to_numpy() #TZ
    crystal_qz_energy = data["QZ"].astype(float).to_numpy() #QZ
    
    print(crystal_dz_energy)
    print(crystal_tz_energy)
    print(crystal_qz_energy)

    # molecular energies with counterpoise correction
    mol_energy = molecular_energies[func]
    mol_dz_energy = mol_energy[0]
    mol_tz_energy = mol_energy[1]
    mol_qz_energy = mol_energy[2]
    
    cohesive_energy_dz = (crystal_dz_energy - 4 * mol_dz_energy) * HA_TO_KJ_MOL / 4
    cohesive_energy_tz = (crystal_tz_energy - 4 * mol_tz_energy) * HA_TO_KJ_MOL / 4
    cohesive_energy_qz = (crystal_qz_energy - 4 * mol_qz_energy) * HA_TO_KJ_MOL / 4 
    
    print("     ", ["111      ", "222      ", "333      ", "444      "])
    print("DZ: ", cohesive_energy_dz)
    print("TZ: ", cohesive_energy_tz)
    print("QZ: ", cohesive_energy_qz)
    
# Theoretical best estimate: âˆ’54.58
```
